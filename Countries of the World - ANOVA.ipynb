# Conducting an Anova for Per Capita GDP, by region

## Goal
***Highlighting Per Capita Income Inequality among regions of the world.***
While more sophisticated analysis is possible & necessary, this is a fairly simple starting point.

## Hypotheses
1. ***$H_{0}$: There is no significant difference in per capita GDP between the regions***
2. ***$H_{A}$: There is a significant difference in per capita GDP between the regions***

```python
import scipy.stats as stats


# Create ANOVA backbone table
data = [['Between Groups', '', '', '', '', '', ''], ['Within Groups', '', '', '', '', '', ''], ['Total', '', '', '', '', '', '']] 
anova_table = pd.DataFrame(data, columns = ['Source of Variation', 'SS', 'df', 'MS', 'F', 'P-value', 'F crit']) 
anova_table.set_index('Source of Variation', inplace = True)

# calculate SSTR and update anova table
x_bar = df['GDP ($ per capita)'].mean()
SSTR = df.groupby('Region').count() * (df.groupby('Region').mean() - x_bar)**2
anova_table['SS']['Between Groups'] = SSTR['GDP ($ per capita)'].sum()

# calculate SSE and update anova table
SSE = (df.groupby('Region').count() - 1) * df.groupby('Region').std()**2
anova_table['SS']['Within Groups'] = SSE['GDP ($ per capita)'].sum()

# calculate SSTR and update anova table
SSTR = SSTR['GDP ($ per capita)'].sum() + SSE['GDP ($ per capita)'].sum()
anova_table['SS']['Total'] = SSTR

# update degree of freedom
anova_table['df']['Between Groups'] = df['Region'].nunique() - 1
anova_table['df']['Within Groups'] = df.shape[0] - df['Region'].nunique()
anova_table['df']['Total'] = df.shape[0] - 1

# calculate MS
anova_table['MS'] = anova_table['SS'] / anova_table['df']

# calculate F 
F = anova_table['MS']['Between Groups'] / anova_table['MS']['Within Groups']
anova_table['F']['Between Groups'] = F

# p-value
anova_table['P-value']['Between Groups'] = 1 - stats.f.cdf(F, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])

# F critical 
alpha = 0.05
# possible types "right-tailed, left-tailed, two-tailed"
tail_hypothesis_type = "two-tailed"
if tail_hypothesis_type == "two-tailed":
    alpha /= 2
anova_table['F crit']['Between Groups'] = stats.f.ppf(1-alpha, anova_table['df']['Between Groups'], anova_table['df']['Within Groups'])

# Final ANOVA Table
anova_table
```

***Since the p-value is less than 0.05 (5%), we reject the null hypothesis and conclude there is significant difference in the per capita income between the countries of the world.***
