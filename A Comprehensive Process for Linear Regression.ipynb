# Defining X and y:

```python
X = df.drop(['Price'], axis = 1)
y = df['Price']
```

```python
X
```

# Fitting the Linear Regression using Statsmodel

***Constant will be added separately.***

```python
import statsmodels.api as sm

X = sm.add_constant(df)
X
```

## Fitting the OLS:

```python
results = sm.OLS(y, X).fit()
results.params
```

***Thus, we have the coefficients of the model***

```python
import numpy as np
A = np.identity(len(results.params))
A = A[1:,:]
A
```

## Getting the Model Summary

```python
results.summary()
```

# Insights from the Summary:

## 1. The F-Stat 

*To check for significance for the model as a whole*

***The Hypotheses are given by:***

***$H_{0}$***: *There is no linear relationship between y and X*

***$H_{1}$***: *There is a linear relationship between y and X*

```python
results.f_test(A)
```

***The p-value for the f-stat is 0.0 - which is significant at 95% confidence (p < 0.05), 97.5% confidence (p < 0.025) and 99% confidence (p < 0.01). Thus, we reject the null hypothesis and conclude that there is a linear relationship between y and X***

## 2. The t-test

*To check for significance of each individual feature*

***The Hypothesis for each feature is given by:***

***$H_{0}$***: *There is no linear relationship between the particular feature and y (i.e. its coefficient is zero or close to it)*

***$H_{1}$***: *There is a linear relationship between the particular feature and y (i.e. its coefficient is far enough away from zero for it to affect y)*


***More formally, for a feature $X_{j}$, (where j = 1, 2, ...p),***

***$H_{0}$***: *$\beta_{j} = 0$*

***$H_{1}$***: *$\beta_{j} \neq 0$*


### From the Model Summary,

```python
results.t_test(A)
```

***Where,***

1. **c0** = Price
2. **c1** = Bedroom
3. **c2** = Space
4. **c3** = Room
5. **c4** = Lot
6. **c5** = Tax
7. **c6** = Bathroom
8. **c7** = Garage
9. **c8** = Condition

Since Price is the dependent variable itself, we look at the p - values of the other factors. Tax has a p-value of 0.000 - which is the only value less than 5%, and thus, significant.

## 3. The Durbin-Watson test for Autocorrelation

The Durbin - Watson test checks for autocorrelation among the residuals. The value of the stat ranges between 0 and 4 - and a value of two indicates no autocorrelation. A value of less than two indicates positive autocorrelation, while a number above two indicates negative autocorrelation.

### From the Model Summary,

```python
results.summary()
```

***The D-W test stat is 1.843, which is close enough to 2 to conclude that there is no (or very little at most) autocorrelation in the data***

## 4. The Jarque-Bera test of Normality

The Jarque-Bera Test determines if a set of data values follows the normal distribution based on the data’s skewness and kurtosis. Here, our null hypothesis is that the data follows a normal distribution, while the alternative is that it does not. 

### From the data,

```python
from scipy.stats import jarque_bera
jarque_bera(df)
```

***The p-value is 0.0, which is significant at 95% confidence (p < 0.05), 97.5% confidence (p < 0.75) and 99% confidence (p < 0.01). Thus, the data does not follow the normal distribution.***

# Improving the Model: Gradient Boosting

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, ensemble
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data & defining params
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=13
)

params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}
```

```python
 # fitting the model:
    
reg = ensemble.GradientBoostingRegressor(**params)
reg.fit(X_train, y_train)
```

```python
# making predictions

pred = reg.predict(X_test)
pred
```

```python
#MAPE:

MAPE = mean_absolute_percentage_error(y_test, pred)
MAPE
```

***That is, MAPE = 0.11%***
